```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calcular_sucesion(A):
    s1 = 0
    sk = s1 
    n = 100
    serie = []
    serie.append(sk)
    while len(serie) < n:
        x = np.random.rand(3)
        x_norm = np.linalg.norm(x, ord=2)
        if x_norm > 1:
            continue
        
        norm_A = np.linalg.norm(A@x) / x_norm
        sk = (max(sk,norm_A))
        serie.append(sk)

    return serie

A = np.array([[1,2,3],[2,3,4],[6,7,8]])
serie = calcular_sucesion(A)

plt.plot(serie)
plt.hlines(y=np.linalg.norm(A, ord=2), xmin=0, xmax=100, colors="r")
plt.show()
```
